Вычислить A + B, A - B
 
$\\
A = \left(
\begin{array}{cccc}
1 & -2 & 4\\
2 & 0 & -1\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
5 & 2 & 3\\
4 & 6 & 2\\
\end{array}
\right)$

```python
A = np.array([[1, -2, 4],
              [2, 0, -1]])
B = np.array([[5, 2, 3],
              [4, 6, 2]])
```

```python
A + B
```

```python
A - B
```

Найти матрицу С = A - 3B
 
$\\
A = \left(
\begin{array}{cccc}
1 & 2 \\
2 & -1\\
3 & 0\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
-1 & 1 \\
1 & 2\\
0 & 0\\
\end{array}
\right)$

```python
A = np.array([[1, 2],
              [2, -1],
              [3, 0]])
B = np.array([[-1, 1],
              [1, 2],
              [0, 0]])
```

```python
A - 3 * B
```

Вычислить С = AB
 
$\\
A = \left(
\begin{array}{cccc}
1 & -1 \\
2 & 0\\
3 & 0\\
\end{array}
\right), B = \left(
\begin{array}{cccc}
1 & 1 \\
2 & 0\\
\end{array}
\right)$

```python
A = np.array([[1, -1],
              [2, 0],
              [3, 0]])
B = np.array([[1, 1],
              [2, 0]])
```

```python
A @ B
```

```python
np.dot(A, B)
```

Вычислить матрицу, обратную матрице А. И показать, что она действительно является обратной
 
$\\
A = \left(
\begin{array}{cccc}
3 & 3 & -1\\
4 & 1 & 3\\
1 & -2 & -2\\
\end{array}
\right)$

```python
A = np.array([[3, 3, -1],
              [4, 1, 3],
              [1, -2, -2]])
```

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
(A @ A_inv).round()
```

```python
np.dot(A, A_inv).round()
```
