#### 1. Выделить полный квадрат

$x^2 + 4x + 6 = x^2 + 2 * x * 2 + 2^2 - 2^2 + 6 = (x + 2)^2 - 2^2 + 6 = (x + 2)^2 + 2$

$x^2 + 9 = x^2 + 2 * x * 0 + 0^2 - 0^2 + 9 = (x + 0)^2 - 0^2 + 9 = (x + 0)^2 + 9$

#### 2. Упростить выражение, используя логарифмирование

$y = x^{\sin x}, x > 0$

$\ln y = \ln x^{\sin x} \Rightarrow \ln y = \sin x \ln x$

#### 3. Определить вид функции (чётная / нечётная / общего вида)

```python
import numpy as np
import matplotlib.pyplot as plt
```

$y = 3x + x^3$

$f(x) = 3x + x^3$

$f(-x) = -3x + (-x^3) = -3x - x^3 = -(3x + x^3) = -f(x)$

Функция $y = 3x + x^3$ нечётная

```python
x = np.arange(-10, 10, 1)
plt.plot(x, 3 * x + x ** 3, 'red')
plt.title('$y = 3x + x^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show();
```

$y = 4^x + 4^{-x}$

$f(x) = 4^x + 4^{-x}$

$f(-x) = 4^{-x} + 4^{-(-x)} = 4^{-x} + 4^x = f(x)$

Функция $y = 4^x + 4^{-x}$ чётная

```python
x = np.arange(-10.0, 10.0, 1.0)
plt.plot(x, 4 ** x + 4 ** -x, 'red')
plt.title('$y = 4^x + 4^{-x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show();
```

$y = (x - 1)^3$

$f(x) = (x - 1)^3$

$f(-x) = (-x - 1)^3 = -x^3 + 3x(-x - 1) - 1 = -x^3 - 3x^2 - 3x - 1$

Функция $y = (x - 1)^3$ общего вида

```python
x = np.arange(-10, 10, 1)
plt.plot(x, (x - 1) ** 3, 'red')
plt.title('$y = (x - 1)^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show();
```
